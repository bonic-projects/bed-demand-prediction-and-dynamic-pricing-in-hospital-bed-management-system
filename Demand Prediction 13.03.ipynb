```python
import warnings
warnings.filterwarnings("ignore")

import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
```

```python
def sum(cols):
    dis=cols[0]
    occ=cols[1]
    tbed=cols[2]
    adm=cols[3]
    ls=cols[4]
    ed=cols[5]
    if dis==0:
        bp=tbed-occ-ed
        return (math.floor(bp))
    else:
        tr=(dis/occ)*100
        bp=tbed-occ+((adm*ls)/tr)-ed 
        return (math.floor(bp)) 

    
# ed is the bed given
```

```python
df["Bed Aailability"]=df[["Discharges","Bed_Occupancy","Hospital_Capacity","Admissions","Length_of_Stay","ED_Traffic"]].apply(sum,axis=1)
```

```python
df.head()
```

```python
df=df.drop(["Patient_Severity","Scheduled_Procedures","Seasonal_Trends","Public_Health_Indicators","Hospital_Policies"],axis=1)
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    train=model.score(xtrain,ytrain)
    test=model.score(xtest,ytest)
    print(f"Training Accuracy:{train}\nTesting Accuracy:{test}\n\n")
    return model
LR=mymodel(LinearRegression())
```

```python
DT=mymodel(DecisionTreeRegressor())
```

```python
def Demand():
    ID=int(input("Enter Hospital ID:"))
    dis=int(input("Enter Discharges:"))
    occ=int(input("Enter Occupancy:"))
    tbed=int(input("Enter Total Bed:"))
    adm=int(input("Enter Admissions:"))
    ls=float(input("Enter Length of Stay:"))
    ed=int(input("Enter ED_Traffic:"))
    
    pred=DT.predict([[dis,occ,tbed,adm,ls,ed]])
    print("The Bed Availability is:",pred)
    print("------------------------------")
    if pred>=tbed:
        print("There is bed demand.")
    else:
        print("There is no bed demand.")
```
